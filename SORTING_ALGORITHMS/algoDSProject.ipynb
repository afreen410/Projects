```python
#Insertion Sort Program
#worst case time complexity - O(n^2)
#best case time complexity - O(n)

import time
from random import sample

#insertion sort program definition
def insertionSort(A):
    for i in range(1,len(A)):
        value = A[i]
        j=i
         #iterating over the unsorted elements and placing them in correct position by shifted sorted elements to the right
        while j>0 and A[j-1]>value:    
            A[j]=A[j-1]
            j=j-1
        A[j]=value

        
sequence=sample(range(1,1000),50)     #50 random numbers generated and passed to insertionSort function
s=time.time()
insertionSort(sequence)
e=time.time()
print('Sorted Sequence-Insertion Sort')
print(sequence)
print('Execution Time :',e-s)


```

```python
# Merge Sort Program

#worst case time complexity is O(logn)
#average case time complexity is O(logn)
from random import sample
import time

def mergeSort(array):

   if len(array)>1:
       mid = len(array)//2
       left = array[:mid]
       right = array[mid:]

       #recursive calls to the left and right array divided based on mid value
       mergeSort(left)        
       mergeSort(right)

       i=0
       j=0
       k=0
        
       #merging the arrays by comparing the elements from the sorted left and right array
       while i < len(left) and j < len(right):
           if left[i] < right[j]:
               array[k]=left[i]
               i=i+1
           else:
               array[k]=right[j]
               j=j+1
           k=k+1

       # merging the remaining elements from left and right array 
       while i < len(left):
           array[k]=left[i]
           i=i+1
           k=k+1

       while j < len(right):
           array[k]=right[j]
           j=j+1
           k=k+1


            
sequence=sample(range(1,1000),50)     #50 random numbers generated and passed to mergeSort function
s=time.time()
mergeSort(sequence)
e=time.time()
print('Sorted Sequence-Merge Sort')
print(sequence)
print('Execution Time :',e-s)



```

```python
from random import sample

def heapInsert(A,e):       #inserting one element at a time and maintaining the min heap property
    A.append(e)
    n=len(A)
    i=n-1
    while i>1 and A[i//2]>A[i] :
        A[i//2],A[i]=A[i],A[i//2]
        i=i//2
def removeMin(A,x):    #remove root element which is the min element 
    n=len(A)
    n=n-x+1
    temp=A[1]
    A[1]=A[n-1]
    n=n-1
    i=1
    while i<n :
        if 2*i+1<=n:
            if A[i]<=A[2*i] and A[i]<=A[2*i+1] :
                return temp
            else:
                if A[2*i]<A[2*i+1]:
                    j=2*i
                else:
                    j=2*i+1
                
                A[i],A[j]=A[j],A[i]
                i=j
        else:
            if 2*i<=n:
                if A[i]>A[2*i]:
                    A[i],A[2*i]=A[2*i],A[i]
            return temp
    return temp
def HeapSort(B):    #call to heap sort with an input of random numbers
    A=[]
    A.append(None)
    for i in B:
        heapInsert(A,i)
    R=[]  
    for i in range(1,len(A)):
        j=removeMin(A,i)
        R.append(j)
#     print('Sorted Sequence-Heap Sort')
#     print(R)
    
            
sequence=sample(range(1,1000),50)     #50 random numbers generated and passed to mergeSort function
s=time.time()
HeapSort(sequence)
e=time.time()
print('Execution Time :',e-s)
```

```python
#In place Quick sort Program : 
import time
from random import sample
import random

def quickSort(A):
    quick_sort(A,0,len(A)-1)

def quick_sort(A,start,end):
    if(start<end):
        pindex = partition(A,start,end)  #call to partition function to get partition index
        quick_sort(A,start,pindex - 1) #call to quick sort function recursively on divided array
        quick_sort(A,pindex + 1,end)
def partition(A,start,end):
    x=random.randint(start,end)   #randomly select a pivot element
    pivot = A[x]
    pindex=start
    for i in range(start,end+1): #looping over the array and swapping till all elements on left of pivot are lesser than it
        if(A[i]<pivot):
            A[i],A[pindex]=A[pindex],A[i]
            pindex=pindex+1
    A[pindex],A[x]=A[x],A[pindex]
    return pindex

#50 random numbers passed as input
sequence=sample(range(1,1000),50)
s=time.time()
sequence.sort()
quickSort(sequence)
e=time.time()
print('Sorted Sequence : In-place quick sort')
print(sequence)
print('Execution Time:',e-s)


```

```python
## Modified Quick Sort for median of 3 elements 

from random import sample
import time


def quickSortModified(arr):
    quicksortMedian(arr,0,len(arr)-1)
    
def medianofThree(arr, left, right):   #finds median of first,middle and last element
    centre = (left + right) // 2
    if (arr[left] > arr[centre]):
        arr[left], arr[centre] = arr[centre], arr[left]
    if (arr[centre] > arr[right]):
        arr[centre], arr[right] = arr[right], arr[centre]
    if (arr[left] > arr[right]):
        arr[left], arr[right] = arr[centre], arr[right]
    arr[centre], arr[right - 1] = arr[right - 1], arr[centre]
    return arr[right - 1]


def quicksortMedian(arr,l,r):
    if l+10<=r:
        pivot = medianofThree(arr,l,r)  #pivot is selected by finding the median of first,middle and last element
        i = l
        j = r-2
        while(i<j):
            while arr[i] < pivot:
                i+=1
            while arr[j] > pivot:
                j-=1
            if(i<j):
                arr[i],arr[j] = arr[j],arr[i]
            else:
                break

        arr[i],arr[r-1] = arr[r-1], arr[i]

        quicksortMedian(arr, l, i-1)
        quicksortMedian(arr, i+1, r)
    else:
        InsertionSort(arr,l,r)  #invoking insertion sort algorithm if the size of array is less than or equal to 10

def InsertionSort(arr,l,r):  #insertion sort algorithm
    for j in range(l,r+1):
        i = j-1
        key = arr[j]
        while i>=0 and arr[i]>key:
            arr[i+1] = arr[i]
            i = i-1
        arr[i+1] = key





#50 random numbers generated and passed to quickSortMedian function
s=time.time()
sequence=sample(range(1,1000),50)     
quickSortModified(sequence)
e=time.time()
print('Sorted Sequence:Modified Quick Sort')
print(sequence)
print('Execution Time :',e-s)

```
